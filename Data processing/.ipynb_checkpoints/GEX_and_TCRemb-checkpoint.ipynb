```python
import warnings
warnings.filterwarnings("ignore")
import muon as mu
import numpy as np
import scanpy as sc
import scirpy as ir
from cycler import cycler
from matplotlib import cm as mpl_cm
from matplotlib import pyplot as plt
from mudata import MuData
import anndata
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tcr_emb = sc.read("TCRemb_2cluster.h5ad")
tcr_emb
```

```python
mdata = mu.read("E:/Python code/Machine learning/JupyterNote/Bio/scripy/data/common_GEX_TCR_mdata.h5mu")
mdata
```

```python
mdata.mod['TCR_emb'] = tcr_emb
mdata
```

```python
common_cells = mdata.mod["TCR_emb"].obs.index.intersection(mdata.mod["airr"].obs.index)

# Subset gex to only include these cells
gex_subset = mdata.mod["gex"][common_cells, :].copy()
tcr_subset = mdata.mod["airr"][common_cells, :].copy()
# Create a new MuData object to preserve alignment
mdata_new = mu.MuData({"gex": gex_subset, "airr": tcr_subset, "airr_emb":tcr_emb })

# Copy previous metadata from the old mdata object
mdata_new.obs = mdata.obs.loc[mdata_new.mod["gex"].obs.index]

# Verify the new object
print(mdata_new)
```

```python
mdata_new.mod['airr_emb'].obs['cluster']
```

```python
mdata_new.obs['tcrEmb_cluster'] = mdata_new.mod['airr_emb'].obs['cluster'].astype('category')
```

```python
mu.pl.embedding(mdata_new, basis="gex:umap", color='tcrEmb_cluster')
```
